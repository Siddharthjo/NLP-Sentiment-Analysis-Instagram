## Sentiment Analysis with Stemming and Classifier Comparison ##

Introduction:
This report presents an analysis of sentiment in Instagram text data using natural language processing (NLP) techniques. The objective is to systematically identify and quantify affective states and subjective information within the text data.

Data Preparation:
The dataset comprises descriptions and variety columns. The descriptions serve as the basis for predicting the variety. Initial data exploration involves handling missing values by imputing them with median or mode values.

Text Preprocessing:
Text preprocessing is crucial for NLP tasks. The descriptions are tokenized and transformed into numerical vectors using the CountVectorizer. To enhance the representation of text data, term frequency-inverse document frequency (TF-IDF) transformation is applied. Additionally, stemming, a process to reduce words to their root form, is introduced using the SnowballStemmer from the Natural Language Toolkit (NLTK).

Model Training:
Two classification algorithms are employed for sentiment analysis: Naive Bayes and Support Vector Machine (SVM). For Naive Bayes, both the standard CountVectorizer and the custom StemmedCountVectorizer incorporating stemming are utilized within a pipeline. The Multinomial Naive Bayes classifier is trained, and predictions are made for the test set. Similarly, an SVM classifier is trained using the SGDClassifier, again with and without stemming.

Evaluation and Comparison:
Model performance is evaluated using accuracy, defined as the proportion of correctly predicted labels. The accuracies of the Naive Bayes and SVM classifiers are compared to determine which algorithm yields better results for sentiment analysis on the given dataset. Results are presented in tabular format, showcasing the algorithm names and their corresponding accuracies, facilitating an easy comparison.

Conclusion:
The analysis demonstrates the effectiveness of applying NLP techniques, including text preprocessing and classification algorithms, for sentiment analysis in Instagram text data. The comparison between Naive Bayes and SVM classifiers provides insights into their relative performance, aiding in informed decision-making for future sentiment analysis tasks. ##

```python
df['description	'] = df.drop(['variety'], axis=1)
X = df['description	']
df['variety'] = df['variety']
y = df['variety']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0,shuffle=True )
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(X_train)
X_train_counts.shape
```

```python
# TF-IDF
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
X_train_tfidf.shape
```

```python
X_train_tfidf
```

**Naive Bayes Algorithm**

```python
# Machine Learning
# Training Naive Bayes (NB) classifier on training data.
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(X_train_tfidf, y_train)
```

```python
# Building a pipeline: We can write less code and do all of the above, by building a pipeline as follows:
# The names ‘vect’ , ‘tfidf’ and ‘clf’ are arbitrary but will be used later.
# We will be using the 'text_clf' going forward.
from sklearn.pipeline import Pipeline

text_clf = Pipeline([('vect', CountVectorizer()), ('tfidf', TfidfTransformer()), ('clf', MultinomialNB())])

text_clf = text_clf.fit(X_train, y_train)
```

```python
text_clf
```

```python
import numpy as np
predicted = text_clf.predict(X_test)
np.mean(predicted == y_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_dt =accuracy_score(y_test,predicted)
print("Accuracy: {}".format(accuracy_dt))
```

**support vector machine**

```python
from sklearn.linear_model import SGDClassifier
text_clf_svm = Pipeline([('vect', CountVectorizer()), ('tfidf', TfidfTransformer()),
                         ('clf-svm', SGDClassifier(loss='hinge', penalty='l2',alpha=1e-3, max_iter=5, random_state=42))])
```

```python
text_clf_svm = text_clf_svm.fit(X_train,y_train)
```

```python
predicted_svm = text_clf_svm.predict(X_test)
```

```python
accuracy_svc = np.mean(predicted_svm == y_test)
accuracy_svc
```

```python
# Stemming Code

#import nltk
#nltk.download()

#from nltk.stem.snowball import SnowballStemmer
#stemmer = SnowballStemmer("english", ignore_stopwords=True)

class StemmedCountVectorizer(CountVectorizer):
    def build_analyzer(self):
        analyzer = super(StemmedCountVectorizer, self).build_analyzer()
        return lambda doc: ([stemmer.stem(w) for w in analyzer(doc)])
    
stemmed_count_vect = StemmedCountVectorizer(stop_words='english')
```
